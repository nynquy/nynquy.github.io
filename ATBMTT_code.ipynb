```python
def right_rotate(value, count):
    return ((value >> count) | (value << (32 - count))) & 0xffffffff
```

```python
def is_prime(p):
    if p < 2:
        return False
    for i in range(2, int(p**0.5) + 1):
        if p % i == 0:
            return False
    return True
```

```python
#Thuật toán Oclip mở rộng
def extended_euclidean_algorithm(a, b):
    if b == 0:
        return a, 1, 0

    gcd, x1, y1 = extended_euclidean_algorithm(b, a % b)
    x = y1
    y = x1 - (a // b) * y1

    return gcd, x, y
```

```python
def mod_inverse(a, m):
    gcd, x, y = extended_euclidean_algorithm(a, m)
    return x % m
```

```python
def find_gcd(a, b): # Hàm tìm UCLN
    while b != 0:
        a, b = b, a % b
    return a
```

```python
#Chia a^b mod c
def mod(alpha,a,p):
    res = 1
    alpha = alpha % p
    while a>0:
        if a%2==1:
            res = (res * alpha) % p
        a = a // 2
        alpha = (alpha * alpha) % p
    return res
```
